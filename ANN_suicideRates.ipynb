```python
# label encoding the data 
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder() 
x['country']= le.fit_transform(x['country']) 
x['sex']= le.fit_transform(x['sex'])
x['age']= le.fit_transform(x['age'])
```

```python
plt.bar(dataset['sex'], y)
plt.xlabel('sex')
plt.ylabel('suicides/100k pop')
plt.title('male vs female suicide rates')
plt.show()
```

```python
plt.bar(dataset['country'], y)
plt.xlabel('country')
plt.ylabel('suicides/100k pop')
plt.title('country suicide rates')
plt.show()
```

```python
plt.barh(dataset["age"], y)
plt.xlabel('Age')
plt.ylabel('suicides/100k pop')
plt.title('Age Group VS Suicide Rates')
plt.show()
```

```python
plt.plot(x["gdp_per_capita ($)"], y, 'o', color='blue');
plt.xlabel('GDP per Capita ($)')
plt.ylabel('Suicide Rate')
plt.title('Suicide Rates vs GDP of the country')
plt.show()
```
