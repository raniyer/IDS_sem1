# Data Cleaning

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('SuicideBombing.csv', encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
df.shape
```

```python
# convert to lower case
df['City'] = df['City'].str.lower()
df['Province'] = df['Province'].str.lower()
# remove trailing white spaces
df['City'] = df['City'].str.strip()
df['Province'] = df['Province'].str.strip()
```

```python
df = df.drop(["S#", "Islamic Date", "Time","Targeted Sect if any","Explosive Weight (max)","Influencing Event/Event", "Temperature(F)"], axis=1)
```

```python
df.isnull().sum()
```

```python
df["Date"]=df["Date"].str.replace("-"," ")
day_name=df.Date.str.split(' ').str[0]
month=df.Date.str.split(' ').str[1]
day=df.Date.str.split(' ').str[2]
year=df.Date.str.split(" ").str[3]
df["Day_name"]=day_name
df["Month"]=month
df["Day"]=day
df["Year"]=year
```

```python
df["Injured Max"]=df["Injured Max"].convert_objects(convert_numeric=True)
```

```python
df['Killed Max'] = df['Killed Max'].fillna(df['Killed Max'].mean())
df['Killed Min'] = df['Killed Min'].fillna(df['Killed Min'].mean())
df['Injured Max'] = df['Injured Max'].fillna(df['Injured Max'].mean())
df['Injured Min'] = df['Injured Min'].fillna(df['Injured Min'].mean())
df['Temperature(C)'] = df['Temperature(C)'].fillna(df['Temperature(C)'].mean())
#deriving columns
df["Killed Avg"]=(df["Killed Max"]+df["Killed Min"])/2
df["Injured Avg"]=(df["Injured Max"]+df["Injured Min"]/2)
#Filling the missing values No. of Suicide Blasts with 1
df["No. of Suicide Blasts"]=df["No. of Suicide Blasts"].fillna("1")
```

```python
df['is Holiday'] = [1 if x ==np.NaN else 0 for x in df['Holiday Type']] 
df = df.drop(["Holiday Type", "Blast Day Type"], axis=1)
```

```python
df['Location'] = df['Location'].fillna(df['Location'].mode())
df = df.sort_values("Location", axis = 0, ascending = True, inplace = False, na_position ='last')
df['Location Category'].replace(to_replace=[None], value=np.nan, inplace=True)
df['Location Category'].fillna(method='ffill', inplace=True)
df['Location Sensitivity'].fillna(method='ffill', inplace=True)
df['Open/Closed Space'].fillna(method='ffill', inplace=True)
df['Longitude'].fillna(method='ffill', inplace=True)
df['Hospital Names'].fillna(method='ffill', inplace=True)
df['Target Type'].fillna(method='ffill', inplace=True)
```
